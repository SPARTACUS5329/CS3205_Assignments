A. 3. List the different MIME types observed in the requests.

```python
import json
from haralyzer import HarParser

def analyze_mimetypes(har_file):
    with open(har_file, 'r', encoding='utf-8') as f:
        har_parser = HarParser(json.loads(f.read()))

    mime_types = set()

    for page in har_parser.pages:
        for entry in page.entries:
            mime_types.add(entry.response.mimeType)

    return mime_types

har_files = ['deccan.har', 'jagran.har', 'mit.har', 'usach.har', 'sinu.har']

for har_file in har_files:
    print(f"Analysis for {har_file}:")
    mime_types = analyze_mimetypes(har_file)
    print("MIME types observed:")
    for mime_type in mime_types:
        print(mime_type)
    print()
```

A. 4. What fraction of GET requests correspond to: (a) Images, (b) javascript and
(c) HTML and CSS

```python
import json
from haralyzer import HarPage

# Dictionary mapping HAR files to their respective page numbers
har_pages = {
    'jagran.har': 'page_1',
    'deccan.har': 'page_2',
    'mit.har': 'page_3',
    'usach.har': 'page_4',
    'sinu.har': 'page_5'
}

# Function to calculate the fraction of GET requests for each content type category
def analyze_har(har_file, page_number):
    with open(har_file, 'r', encoding='utf-8') as f:
        har_data = json.loads(f.read())
    
    # Create a HarPage object for the specified page number
    har_page = HarPage(page_number, har_data=har_data)
    
    # Filter entries based on content type
    image_entries = har_page.filter_entries(content_type='image.*', request_type='GET')
    js_entries = har_page.filter_entries(content_type='application/javascript', request_type='GET')
    html_entries = har_page.filter_entries(content_type='text/html', request_type='GET')
    css_entries = har_page.filter_entries(content_type='text/css', request_type='GET')
    
    # Calculate the fractions
    total_get_requests = len(har_page.filter_entries(request_type='GET'))
    image_fraction = len(image_entries) / total_get_requests
    js_fraction = len(js_entries) / total_get_requests
    html_fraction = len(html_entries) / total_get_requests
    css_fraction = len(css_entries) / total_get_requests
    
    return {
        'image_fraction': image_fraction,
        'js_fraction': js_fraction,
        'html_fraction': html_fraction,
        'css_fraction': css_fraction
    }

# Perform analysis for each HAR file
analysis_results = {}
for har_file, page_number in har_pages.items():
    analysis_results[har_file] = analyze_har(har_file, page_number)

# Display the results
for har_file, results in analysis_results.items():
    print(f"Analysis for {har_file}:")
    print(f"Fraction of GET requests for images: {results['image_fraction']:.2f}")
    print(f"Fraction of GET requests for javascript: {results['js_fraction']:.2f}")
    print(f"Fraction of GET requests for HTML: {results['html_fraction']:.2f}")
    print(f"Fraction of GET requests for CSS: {results['css_fraction']:.2f}")
    print()
```

A. 5. What is the total size of the assets downloaded as a part of the responses?
What fraction of the size constitute of the images? What is the mean and
median size of the images?

```python
import json
from haralyzer import HarPage

# Dictionary mapping HAR files to their respective pages
har_files_pages = {
    "deccan.har": "page_2",
    "jagran.har": "page_1",
    "mit.har": "page_3",
    "usach.har": "page_4",
    "sinu.har": "page_5"
}

# Iterate over each HAR file
for har_file, page_name in har_files_pages.items():
    with open(har_file, 'r', encoding='utf-8') as f:
        # Load HAR data and create HarPage object for the specified page
        har_data = json.loads(f.read())
        har_page = HarPage(page_name, har_data=har_data)
        
        # Initialize variables
        total_asset_size = 0
        total_image_size = 0
        image_sizes = []
        
        # Iterate over entries and calculate sizes
        for entry in har_page.entries:
            if entry.response is not None and entry.response.contentSize is not None:
                total_asset_size += entry.response.contentSize
                if entry.response.contentType and entry.response.contentType.startswith('image'):
                    total_image_size += entry.response.contentSize
                    image_sizes.append(entry.response.contentSize)
        
        # Calculate fraction of image size
        fraction_images = total_image_size / total_asset_size if total_asset_size > 0 else 0
        
        # Calculate mean size of images
        mean_image_size = sum(image_sizes) / len(image_sizes) if len(image_sizes) > 0 else 0
        
        # Calculate median size of images
        sorted_image_sizes = sorted(image_sizes)
        median_image_size = sorted_image_sizes[len(sorted_image_sizes) // 2] if sorted_image_sizes else 0
        
        # Print the result for the current HAR file
        print(f"HAR File: {har_file}")
        print(f"Total Asset Size: {total_asset_size}Kb")
        print(f"Total Image Size: {total_image_size}Kb")
        print(f"Fraction of Image Size: {fraction_images}")
        print(f"Mean Image Size: {mean_image_size}Kb")
        print(f"Median Image Size: {median_image_size}Kb")
        print()
```
